# Key Elements of Descriptive Statistics

### Measures Of Central Tendency

* Mean
* Median
* Mode

### Measures Of Spread

* Range
* Outliers
* Interquantile Range
* Variance
* Standard deviation

### Dependence

* Correlation

# we will begin our initial descriptive statistical analysis.

For our analysis, we will chosen 'Population in thousands (2017)' column. We can perform the similar operations on other columns as well. We will get the measure of central tendency and measure of spread of the data.

```python
# mean

# Manually
data['Population in thousands (2017)'].sum()/len(data['Population in thousands (2017)'])
```

```python
#mean - we will check the mean of the Population in thousands (2017)
data['Population in thousands (2017)'].mean()
```

```python
#median - we will check the median of the Population in thousands (2017)
data['Population in thousands (2017)'].median()
```

```python
#mode - we will check the mode of the Population in thousands (2017)
data['Population in thousands (2017)'].mode()
```

```python
data['Population in thousands (2017)'].mode()[3] #if you want to excess one value
```

```python
# Range : max - min

a = data['Population in thousands (2017)'].max()
b = data['Population in thousands (2017)'].min()

print("Max value : ", a)
print("Min value: ", b)
```

```python
range = a - b
range
```

```python
#variance
variance = data['Population in thousands (2017)'].var()
```

```python
#standard Deviation
data['Population in thousands (2017)'].std()
```

```python
# Manual Std
import math
std_deviation = math.sqrt(variance)
std_deviation
```

```python
# Percentile

#75, 50 and 25 percentile
import numpy as np

sev = np.percentile(data['Population in thousands (2017)'], 75)
fif = np.percentile(data['Population in thousands (2017)'], 50)
twe = np.percentile(data['Population in thousands (2017)'], 25)

print("75th Percentile is : ", sev)
print("50th Percentile is : " , fif)
print("25th Percentile is : ", twe)
```

```python
# Quantile

# Quartile

q1 = data['Population in thousands (2017)'].quantile(0.25)
q2 = data['Population in thousands (2017)'].quantile(0.50)
q3 = data['Population in thousands (2017)'].quantile(0.75)

print(q1,q2,q3)
```

```python
# IQR

iqr = q3 - q1
iqr
```

```python
# correlation
# positive, negative, zero

data["Population in thousands (2017)"].corr(data["Population density (per km2, 2017)"])
```
