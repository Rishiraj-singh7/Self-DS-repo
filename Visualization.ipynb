### line chart

```python
plt.plot(x,y)
plt.show()
```

```python
# customize the line

fig = plt.figure(figsize = (5,5))
plt.plot(x,y, linewidth = 2, linestyle = "--", color = 'red', alpha = 1)      # multiple stryle, :, -, -- # alpha = tranasparency (0-1)
plt.title("Line chart demo")
plt.xlabel("x-lable")
plt.ylabel("y-lable")
plt.legend(["line 1"], loc = "best")
plt.show()
```

```python
# plot 2 charts
x = np.arange(0,10)
y1 = 2*x + 5
y2 = 10*x + 3 
```

```python
# create two or more plots in a figure

fig = plt.figure(figsize = (5,5))
plt.subplot(1,2,1)#no of rows, no of columns, index/subplot
plt.plot(x,y1)
plt.title("graph 1")

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title("graph 2")
plt.show()
```

```python
# create two or more plots in a figure

fig = plt.figure(figsize = (5,5))
plt.subplot(2,1,1)#no of rows, no of columns, index/subplot
plt.plot(x,y1)
plt.title("graph 1")

plt.subplot(2,1,2)
plt.plot(x,y2)
plt.title("graph 2")
plt.show()
```

```python
#create data
x = np.arange(0,10)

y1 = 2*x + 5
y2 = 10*x + 3
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["Line 1","Line 2"], loc = "best")
plt.show()
```

```python
#Always careful while custamization that you can't lose data and its perpective,,, don't do until they show some meaning
plt.plot(x,y1, marker = 'o')#1,2,4,o,d,s,p,g,<,>,^,x
plt.plot(x,y2, marker = '>')
plt.legend(["Line 1","Line 2"], loc = "best")
plt.show()
# when ever you see trend or study with repect of time(trend) we choose line chart
```

```python
#bar chart = we use to know the result in breaks or period pf time, use b-chart(catogarical value (means no.fixed value))
data = {"NW": 357,"NE": 175,"SW": 254,"SE": 657}
data
```

```python
region = list(data.keys())
counts = list(data.values())
print(region)
print(counts)
#converting to list to plot it on bar-chart
```

```python
# plot the bar chart - catogarical variable vs aggregated value
plt.bar(region, counts)
plt.show()
```

```python
#Customization of the bar chart
plt.bar(region, counts, color="orange", width = 0.5)
plt.title("Regionwise Population")
plt.xlabel('Region')
plt.ylabel('Population')
#make text on bar
for index,value in enumerate(counts):
    plt.text(index, value -50, str(value))
plt.show()
```

```python
plt.barh(region, counts, color="black", height = 0.4)
plt.title("Regionwise Population")
plt.xlabel('Region')
plt.ylabel('Population')
plt.show()
```

```python
# change the color of individual bars
bars = plt.bar(region, counts, width = 0.4)
bars[0].set_color('green')
bars[1].set_color('grey')
bars[2].set_color('red')
bars[3].set_color('#001')
```

```python
# Histogram
age = [38,22,40,37,86,34,85,53,24,30,28,85,82,62,23,96,57,59,86,78]
plt.hist(age, bins = [0,20,40,60,80,100], color = 'green')
plt.grid(True)
plt.show()
```

```python
#Box plot chart
age = [38,22,40,37,86,34,85,53,24,30,28,85,82,62,23,206,57,59,86,78]
plt.boxplot(age)
plt.grid(True, color = 'green', linewidth = 0.5, linestyle = ":")
plt.show
# Bottom Whisker => min value
# Bottom line of the box  => 25%ile
# orange => 50%ile/Median
# top line of the box => 75%ile
# top Whisker => max value
# Dot => outlier
```
