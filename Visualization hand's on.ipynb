```python
fig = plt.subplots(2,2, figsize = (30,20))

x1 = df['hp']
x2 = df['mpg']
x3 = df['weightlbs']
x4 = df['time-to-60']

plt.subplot(2,2,1)
sns.barplot(x = x1, y = 'cylinders', data = df)

plt.subplot(2,2,2)
sns.barplot(x = x2, y = 'cylinders', data = df)

plt.subplot(2,2,3)
sns.barplot(x = x3, y = 'cylinders', data = df)

plt.subplot(2,2,4)
sns.barplot(x = x4, y = 'cylinders', data = df)

plt.show()
```
