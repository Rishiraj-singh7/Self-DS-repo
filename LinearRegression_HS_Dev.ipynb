# Imputing the data for the missing records

```python
for col_name in col_list:
    if df[col_name].dtypes == 'object':
        df[col_name] = df[col_name].fillna(df[col_name].mode()[0])
    else:
        df[col_name] = df[col_name].fillna(df[col_name].mean())
```

# * Treatment of Outliers

```python
df.describe()
```

```python
# BMI
Q1 = df.bmi.quantile(0.25)
Q3 = df.bmi.quantile(0.75)
Q1, Q3
```

```python
# IQR -> Interquartile Range
IQR = Q3 - Q1
IQR
```

```python
# upper limit and lower limit
# Lowerlimit
LL = Q1 - (1.5 * IQR)
UL = Q3 + (1.5 * IQR)
LL, UL
```

```python
plt.boxplot(df['bmi'])
plt.show()
```

```python
# treatment phase
df = df[(df.bmi >= LL) & (df.bmi <= UL)]
```

```python
plt.boxplot(df['bmi'])
plt.show()
```

```python
Q1 = df.past_consultations.quantile(0.25)
Q3 = df.past_consultations.quantile(0.75)
IQR = Q3 - Q1
LL = Q1 - (1.5 * IQR)
UU = Q3 + (1.5 * IQR)
df = df[(df.past_consultations >= LL) & (df.past_consultations <= UL)]
```

```python
plt.boxplot(df['past_consultations'])
plt.show()
```

 UL = Q3 + 1.5 * IQR

 LL = Q1 - 1.5 * IQR

```python
Q1 = df.Hospital_expenditure.quantile(0.25)
Q3 = df.Hospital_expenditure.quantile(0.75)
IQR = Q3 - Q1
df = df[(df.Hospital_expenditure >= Q1 - 1.5*IQR) & (df.Hospital_expenditure <= Q3 + 1.5*IQR)]

Q1 = df.Anual_Salary.quantile(0.25)
Q3 = df.Anual_Salary.quantile(0.75)
IQR = Q3 - Q1
df = df[(df.Anual_Salary >= Q1 - 1.5*IQR) & (df.Anual_Salary <= Q3 + 1.5*IQR)]
```
